## 3. ***Data Wrangling***

### Data Wrangling Code

Convert Date Columns

````python
# 1. Convert columns using the specific format (Day-MonthName-Year)
date_cols = ['Scheduled Delivery Date', 'Delivered to Client Date']

for col in date_cols:
    # Adding format='%d-%b-%y' tells Pandas: "Expect 13-Jul-06"
    # This stops the Warning!
    df[col] = pd.to_datetime(df[col], format='%d-%b-%y', errors='coerce')

# 2. Calculate the delay as a number
df['Delivery Delay (days)'] = (df['Delivered to Client Date'] - df['Scheduled Delivery Date']).dt.days

# 3. Check if it worked
print("How many rows of data do we have now?")
print(df['Delivery Delay (days)'].count())
````

Convert Numeric Columns

````python
# Remove commas and convert to float
df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'], errors='coerce')
df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'], errors='coerce')
````

Create New Useful Columns

````python
# 1. Convert PO Sent to Vendor Date (Letting Pandas guess the format)
df['PO Sent to Vendor Date'] = pd.to_datetime(df['PO Sent to Vendor Date'], errors='coerce')

# 2. Re-calculate Lead Time
# Lead Time: From the day the PO was sent to the day it was delivered
df['Lead Time (days)'] = (df['Delivered to Client Date'] - df['PO Sent to Vendor Date']).dt.days

# 3. Check the results again
print(df[['Delivery Delay (days)', 'Lead Time (days)']].describe())
````

````python
# Fill missing Lead Time using median (safe for skewed data)
median_lead_time = df['Lead Time (days)'].median()
print(f"Median Lead Time: {median_lead_time} days")

# Clean pandas 3.0 compatible syntax
df['Lead Time (days)'] = df['Lead Time (days)'].fillna(median_lead_time)

# Verify fix
print("Lead Time null count after fix:", df['Lead Time (days)'].isnull().sum())
print(df['Lead Time (days)'].describe())
````

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

H0 (Null Hypothesis): There is no significant difference in average freight cost between Air and non-Air shipments.
H1 (Alternate Hypothesis): Air shipments have a significantly higher average freight cost than non-Air shipments.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy import stats

# Separate Air and Non-Air freight costs, drop nulls
air_cost = df[df['Shipment Mode'] == 'Air']['Freight Cost (USD)'].dropna()
non_air_cost = df[df['Shipment Mode'] != 'Air']['Freight Cost (USD)'].dropna()

# Mann-Whitney U Test (non-parametric, suitable for skewed data)
stat, p_value = stats.mannwhitneyu(air_cost, non_air_cost,
                                    alternative='greater')

print("=" * 50)
print("Hypothesis 1: Air vs Non-Air Freight Cost")
print("=" * 50)
print(f"Mann-Whitney U Statistic: {stat:.4f}")
print(f"P-Value: {p_value:.6f}")
print()
if p_value < 0.05:
    print("Result: REJECT Null Hypothesis")
    print("Conclusion: Air shipments have significantly")
    print("higher freight costs than non-Air shipments.")
else:
    print("Result: FAIL TO REJECT Null Hypothesis")
    print("Conclusion: No significant difference found.")
print("=" * 50)
````

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

H0 (Null Hypothesis): There is no significant difference in delivery delays across different shipment modes (Air, Truck, Air Charter, Ocean).
H1 (Alternate Hypothesis): Delivery delays differ significantly across at least one pair of shipment modes.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy import stats

# Separate delivery delays by shipment mode, drop nulls
air_delay = df[df['Shipment Mode'] == 'Air']['Delivery Delay (days)'].dropna()
truck_delay = df[df['Shipment Mode'] == 'Truck']['Delivery Delay (days)'].dropna()
charter_delay = df[df['Shipment Mode'] == 'Air Charter']['Delivery Delay (days)'].dropna()
ocean_delay = df[df['Shipment Mode'] == 'Ocean']['Delivery Delay (days)'].dropna()

# Kruskal-Wallis Test (non-parametric ANOVA for multiple groups)
stat, p_value = stats.kruskal(air_delay, truck_delay,
                               charter_delay, ocean_delay)

print("=" * 50)
print("Hypothesis 3: Delivery Delay by Shipment Mode")
print("=" * 50)
print(f"Kruskal-Wallis Statistic: {stat:.4f}")
print(f"P-Value: {p_value:.6f}")
print()
if p_value < 0.05:
    print("Result: REJECT Null Hypothesis")
    print("Conclusion: Delivery delays differ significantly")
    print("across shipment modes.")
else:
    print("Result: FAIL TO REJECT Null Hypothesis")
    print("Conclusion: No significant difference in delays")
    print("across shipment modes.")
print("=" * 50)
````

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments

def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    before = df[column].describe()
    df[column] = df[column].clip(lower=lower, upper=upper)
    after = df[column].describe()

    print(f"\n{column}:")
    print(f"  Lower cap: {lower:.2f} | Upper cap: {upper:.2f}")
    print(f"  Max BEFORE: {before['max']:.2f} | Max AFTER: {after['max']:.2f}")

# Apply to key numerical columns
outlier_cols = ['Weight (Kilograms)', 'Freight Cost (USD)',
                'Line Item Value', 'Line Item Insurance (USD)']

print("Outlier Treatment using IQR Capping:")
for col in outlier_cols:
    cap_outliers_iqr(df, col)

print("\nOutlier treatment completed successfully!")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Answer Here.


IQR Capping (Winsorization) technique was used for outlier treatment. This method calculates the Interquartile Range (IQR) and caps values below Q1 - 1.5×IQR and above Q3 + 1.5×IQR. This technique was chosen over removal because removing outliers would result in significant data loss given the large number of extreme values in freight cost, weight, and line item value columns. Capping preserves all records while limiting the influence of extreme values on model training, making it the most appropriate technique for this financial logistics dataset.

### 4. Feature Manipulation & Selection

#### 1. Feature Manipulation

````python
# Manipulate Features to minimize feature correlation and create new features

import numpy as np

# 1. Cost per KG - measures freight efficiency
df['Cost_per_KG'] = (df['Freight Cost (USD)'] /
                     df['Weight (Kilograms)'].replace(0, np.nan))

# 2. Insurance Rate - measures insurance as % of shipment value
df['Insurance_Rate'] = (df['Line Item Insurance (USD)'] /
                        df['Line Item Value'].replace(0, np.nan))

# 3. Value per Unit - measures unit level pricing
df['Value_per_Unit'] = (df['Line Item Value'] /
                        df['Line Item Quantity'].replace(0, np.nan))

# Fill any nulls created by division
df['Cost_per_KG'] = df['Cost_per_KG'].fillna(df['Cost_per_KG'].median())
df['Insurance_Rate'] = df['Insurance_Rate'].fillna(df['Insurance_Rate'].median())
df['Value_per_Unit'] = df['Value_per_Unit'].fillna(df['Value_per_Unit'].median())

print("New features created successfully:")
print(f"  ✓ Cost_per_KG     - Mean: {df['Cost_per_KG'].mean():.2f}")
print(f"  ✓ Insurance_Rate  - Mean: {df['Insurance_Rate'].mean():.4f}")
print(f"  ✓ Value_per_Unit  - Mean: {df['Value_per_Unit'].mean():.2f}")
print(f"\nDataset shape after manipulation: {df.shape}")
````

### 7. Dimesionality Reduction

##### Do you think that dimensionality reduction is needed? Explain Why?

Answer Here.

Yes, dimensionality reduction was considered because the Correlation Heatmap in Chart 14 revealed strong correlations between several features, particularly Line Item Value and Line Item Insurance (0.96) and Weight and Freight Cost (0.74). Highly correlated features introduce multicollinearity which can negatively impact model performance. PCA was applied to identify how many components are needed to explain 95% of the variance, helping reduce redundant information while preserving the most important patterns in the data.

````python
# DImensionality Reduction (If needed)
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Use scaled numerical features
pca_cols = ['Weight (Kilograms)',
            'Freight Cost (USD)',
            'Line Item Value',
            'Line Item Insurance (USD)',
            'Line Item Quantity',
            'Pack Price',
            'Unit Price',
            'Delivery Delay (days)',
            'Lead Time (days)']

# Drop nulls
df_pca = df_scaled[pca_cols].dropna()

# Apply PCA
pca = PCA()
pca.fit(df_pca)

# Cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(10, 5))
plt.plot(range(1, len(cumulative_variance) + 1),
         cumulative_variance, marker='o', color='steelblue')
plt.axhline(y=0.95, color='red', linestyle='--',
            label='95% Variance Threshold')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA - Cumulative Explained Variance', fontsize=14)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("PCA Explained Variance per Component:")
for i, var in enumerate(pca.explained_variance_ratio_):
    print(f"  PC{i+1}: {var*100:.2f}% "
          f"(Cumulative: {cumulative_variance[i]*100:.2f}%)")
````

##### Which dimensionality reduction technique have you used and why? (If dimensionality reduction done on dataset.)

Answer Here.

PCA (Principal Component Analysis) was used for dimensionality reduction. PCA was chosen because it is the most widely used linear dimensionality reduction technique that transforms correlated features into uncorrelated principal components. It maximizes the variance explained by each component, allowing us to retain the most important information while discarding redundant features. This reduces computational complexity and helps prevent overfitting in machine learning models.
